```python
#everything buildup

import pandas as pd

# Assuming possession_stats_df and passing_stats_df are the DataFrames for possession and passing respectively

# Step 1: Convert columns to numeric (if not already)
possession_stats_df['Mis'] = pd.to_numeric(possession_stats_df['Mis'], errors='coerce')
possession_stats_df['Dis'] = pd.to_numeric(possession_stats_df['Dis'], errors='coerce')
possession_stats_df['PrgC'] = pd.to_numeric(possession_stats_df['PrgC'], errors='coerce')

# Calculate Mis+Dis from possession_stats_df
possession_stats_df['Mis+Dis'] = possession_stats_df['Mis'] + possession_stats_df['Dis']

# Merge possession_stats_df with passing_stats_df on 'Player' column
merged_df = possession_stats_df.merge(passing_stats_df[['Player', 'PrgP']], on='Player', how='left')

# Convert 'PrgP' column to numeric (if not already)
merged_df['PrgP'] = pd.to_numeric(merged_df['PrgP'], errors='coerce')

# Convert '90s' column to numeric (if not already)
merged_df['90s'] = pd.to_numeric(merged_df['90s'], errors='coerce')

# Calculate ProgP + ProgC
merged_df['ProgP+ProgC'] = merged_df['PrgP'] + merged_df['PrgC']

# Perform safe division to prevent division by zero errors
merged_df['Score'] = merged_df['ProgP+ProgC'] / merged_df['Mis+Dis'].where(merged_df['Mis+Dis'] != 0, other=1)

# Divide the obtained score by the number of 90s played
merged_df['Buildup_Score'] = merged_df['Score'] / merged_df['90s'].where(merged_df['90s'] != 0, other=1)

# Assign the 'Buildup_Score' to possession_stats_df
possession_stats_df['Buildup_Score'] = merged_df['Buildup_Score']
```

```python
# Filter rows where '90s' is greater than 10 and 'Pos' contains 'MF'
filtered_merged_df = merged_df[(merged_df['90s'] > 10) & (merged_df['Pos'].str.contains('MF', case=False, na=False))]

# Sort the filtered DataFrame by 'Buildup_Score' column in descending order
sorted_filtered_merged_df = filtered_merged_df

from scipy.stats import percentileofscore
# Calculate percentiles for the 'Buildup_Score' column
sorted_filtered_merged_df['Buildup_Score_Percentile'] = sorted_filtered_merged_df['Buildup_Score'].apply(lambda x: percentileofscore(sorted_filtered_merged_df['Buildup_Score'], x))
# Print the first few rows of the sorted and filtered DataFrame
sorted_filtered_merged_df.head()
```
